```python
import pandas as pd
import numpy as np
import os
```

## 1. Access Data

```python
fh = os.path.join("..","..","Data")
files = [f for f in os.listdir(fh) if f.endswith(".csv")]
files
```

## 2. Import and preview data

```python
us = pd.read_csv(os.path.join(fh,"SpotifyUSHits.csv"))
us.head(2)
```

```python
tks = pd.read_csv(os.path.join(fh,"TrackClean.csv"))
tks.head(2)
```

```python
albums = pd.read_csv(os.path.join(fh,"Albums.csv"))
albums.head(2)
```

## 3. Merge Data

```python
# Hits and TrackClean merge on TrackID and id respectively
m1 = us.merge(tks,left_on = 'TrackID',right_on = 'id',how = 'left')
```

```python
#Album and TrackClean merge on album_id
m2 = m1.merge(albums,on = 'album_id',how = 'left')
```

```python
m2.dtypes
```

```python
#So this is all of the hits, track, and album data. You could add artist and track data if you wanted too. 
m2.head()
```

```python
m2['Day']= pd.to_datetime(m2['Day'])
m2['release_date']= pd.to_datetime(m2['release_date'])
```

```python
m2['Days_since_album_release']= m2['Day']-m2['release_date']
m2['Days_since_album_release']= [x.days for x in m2['Days_since_album_release']]
```

```python
m2.head()
```

```python
#What is the oldest song to hit the top charts?
m2[m2['Days_since_album_release']==max(m2['Days_since_album_release'])]
```

```python
#Which labels have the most #1 hits?
m2[m2['Rank']==1]['label'].value_counts()
```
